```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # builds on matplotlib
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#let's import our data
#Windows:
stats = pd.read_csv('C:\\Users\\houci\\Desktop\\Python Programming Course\\Demographic-Data.csv')
```

```python
stats.head()
```

```python
#Rename columns
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',
       'IncomeGroup']
```

```python
stats.head()
```

```python
#Distribution:
vis1 = sns.distplot(stats['InternetUsers'])
```

```python
vis1 = sns.distplot(stats['InternetUsers'], bins = 30)
```

**Interesting Observation**:

Obviously, quite a few countries have 0 to 20% access to the internet and most countries fall between 20% and 70% internet usage and the usage slowly declines from 80%.

```python
#Boxplots:
vis2 = sns.boxplot(data = stats, x ='IncomeGroup', y= 'BirthRate')
```

#### **Insights:**

The boxplot shows the average birth rate of high-income earners is at about 12 percent while the average birth rate for low-income group is at about 38 percent. The upper middle income earners are leaning towards 18 percent and lower middle incomme group have about 28 percent birth rate.

**Note:** We do have some outliers in the dataset.

```python
#Advance Visualization
```

```python
#lm- linear model plot
vis3 = sns.lmplot (data=stats, x='InternetUsers', y = 'BirthRate',fit_reg =False)
```

```python
#let's add some color to the scatterplot
vis3 = sns.lmplot (data=stats, x='InternetUsers', y = 'BirthRate', 
                   fit_reg =False, hue = 'IncomeGroup')

#We change facet grid size to 10
#aspect ratio default is 1
```

####  Interesting Insights:

From the plot we can see that majority of high income and upper middle income group are between 10 to 20 percent birth rate and lower and lower middle income group are above 20 percent. Hence it seems those with with the lower birth rate have high access to the internet and the groups with high birth rate have limited access to internet

**Note:** Correlation does not imply causation
