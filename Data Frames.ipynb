#### Filtering Data Frames

```python
#We filter by rows
```

```python
stats.head()
```

```python
Filter = stats.InternetUsers < 2
```

```python
Filter
```

```python
stats[Filter] #Conceptually this is just like R
```

```python
#Let's do more practice
Filter2 = stats.BirthRate > 40
```

```python
stats[Filter2]
```

```python
#More interesting practice:
stats[stats.BirthRate > 40 and stats.InternetUsers < 2] #Why do we have an error?
#This will not give an error in R
#Python is not a vectorized programming language
# *and* is looking for a single value(True/False) on the left and right objects
```

```python
#Error Again Why?
Filter and Filter2 
#Filter and Filter2 are series objects
```

```python
# &- 
#True- when all conditions are met
Filter & Filter
```

```python
#Output: where both conditions are true
stats[Filter & Filter2]
```

```python
#use () to seperate the objects and run below and prevent errors
stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)]
```

#### Using .at() and .iat()

```python
#Access individual elements
```

```python
stats[2,2] #Error. Why?
```

```python
stats.head()
```

```python
#We need the:
#.at for labels. Note: Integers are treated like labels.
#.iat for integer for integers
```

```python
stats.iat[3,4]
```

```python
stats.iat[0,1]
```

```python
#to get labels
stats.at[2, 'BirthRate']
```

```python
#why do we need .at:
sub10 = stats[::10]
```

```python
sub10
```

```python
#.iat counts from the rows(0123....)
sub10.iat[10,0]
```

```python
#.at looks for the label, 
#in this case label is 10 and CountryName
sub10.at[10, 'CountryName']
```

```python
#.at looks for the label, 
#in this case label is 10 and CountryName
sub10.at[60, 'IncomeGroup']
```
