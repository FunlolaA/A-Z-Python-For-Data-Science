```python
#Let's create our own function
#how can we make it easier? Use a 'For' loop

def myplot(playerlist):
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c='Black', ls='--', marker='s', ms=7, label=name) # player 0
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical') #rotate the axis
    plt.show()
```
